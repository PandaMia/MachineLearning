6. Какое самое популярное женское имя на корабле? Извлеките из полного имени пассажира (колонка Name) его личное имя (First Name).

```python
def first_name(name):
    split_name = name.split('. ')
    if '(' in name:
        split_name = split_name[1].split('(')
    first_name = split_name[1].split(' ')
    return first_name[0]
```

```python
titanic['First_name'] = titanic['Name'].apply(first_name)
```
