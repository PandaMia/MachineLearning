```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
data = pd.read_csv('C:/Users/1/Data/titanic_coursera.csv', index_col='PassengerId')
```

```python
data.shape
```

```python
train = data.loc[:, ['Survived', 'Pclass', 'Fare', 'Age', 'Sex']]
```

```python
train.head()
```

```python
train['Sex'] = (train['Sex'] == 'female').astype('int')
```

```python
train.head()
```

```python
train[train['Age'].isnull()]
```

```python
train.dropna(inplace = True)
```

```python
train.head()
```

```python
labels = train['Survived']
```

```python
train.drop(['Survived'], axis = 1, inplace = True)
```

```python
X = train.values
```

```python
Y = labels.values
```

```python
My_tree = DecisionTreeClassifier()
My_tree.fit(X,  Y)
```

```python
importances = My_tree.feature_importances_
```
